```python
from icdmappings import Mapper
from icdmappings import Validator

mapper = Mapper()
validator = Validator()
```

## ED

### แปลง icd version ที่ผิด จาก 9 เป็น 10

```python
ed_icd9_false = ed[(ed['icd_version'] == 9) & (ed['icd_code'].apply(lambda x: not validator.validate(x, expects='icd9_diagnostic')))]
ed_icd9_false
```

```python
# เปลี่ยน version ที่ผิดจาก ICD 9 ให้เป็น 10
ed.loc[(ed['icd_version'] == 9) & (ed['icd_code'].apply(lambda x: not validator.validate(x, expects='icd9_diagnostic'))),'icd_version'] = 10
```

### Grouping ED ICD9 to chapter

```python
ed_9 = ed[ed['icd_version']==9]
ed_9.head()
```

```python
ed_9["chapter"] = mapper.map(ed_9["icd_code"], source='icd9', target='chapter')
ed_9
```

```python
len(ed_9[ed_9['chapter'].isna()]) # ทุก chapter จาก icd9 map ได้
```

### Grouping ED ICD10 to chapter

```python
ed_10 = ed[ed['icd_version']==10]
ed_10.head()
```

```python
ed_10["chapter"] = mapper.map(ed_10["icd_code"], source='icd10', target='chapter')
ed_10
```

```python
len(ed_10[ed_10['chapter'].isna()]) # มี 3404 ตัวที่ map จาก ICD10 => chapter ไม่ได้
```

```python
ed_10_nullchapter = ed_10[ed_10['chapter'].isna()]
ed_10_nullchapter
```

```python
# ณ ที่นี้จะทำการ drop ทิ้งก่อน
ed_10 = ed_10.drop(index=ed_10_nullchapter.index)
len(ed_10[ed_10['chapter'].isna()])
```

### Grouping ED Result

```python
ed_10.head()
```

```python
ed = pd.concat([ed_9,ed_10], axis=0)
```

```python
c1 = 0
c2 = 0
for e in validator.validate(ipd[ipd['icd_version'] == 10]['icd_code'], expects='icd10_diagnostic'):
    if e == True:
        c1 += 1
    else:
        c2 += 1
print(f'len val true: {c1}\nlen val false: {c2}') # ลองใช้ validator ดูเฉยๆ สุดท้ายถ้า map ได้ก็ไม่น่ามีปัญหา
```
